```python
from pwn import *  # pip install pwntools
#from Crypto.Util.number import  long_to_bytes
from Cryptodome.Util.number import *
import json
import base64
import codecs
import random
import binascii


def json_recv():
    line = r.recvline()
    return json.loads(line.decode())


def json_send(hsh):
    request = json.dumps(hsh).encode()
    r.sendline(request)

def decoder(received):

    if (received["type"] == "rot13"):
        var = codecs.decode(received["encoded"], 'rot_13')
        to_send = {
            "decoded": var
        }
    elif (received["type"] == "hex"):
        var = bytes.fromhex(received["encoded"]).decode('utf-8')
        to_send = {
            "decoded": var
        }
    elif (received["type"] == "base64"):
        #var = base64.b64decode(received["encoded"]).decode('utf-8')
        var= base64.b64decode(received["encoded"]).decode('utf-8')
        #var = var.encode()
        to_send = {
            "decoded": var
        }
    elif (received["type"] == "bigint"):
        var = received["encoded"]
        var = binascii.unhexlify(var.replace("0x","")).decode("utf-8")
        to_send = {
            "decoded": var
        }
    elif (received["type"] == "utf-8"):
        s = ""
        for b in received["encoded"]:
            s = s+chr(b)
        
        to_send = {
            "decoded": s
        }
    else:
        to_send = {
            "decoded": "no"
        }

    return to_send
```

```python
r = remote('socket.cryptohack.org', 13377, level = 'debug')
while (True):
    received = json_recv()

    # print("Received type: ")
    # print(received["type"])
    # print("Received encoded value: ")
    # print(received["encoded"])
    json_send(decoder(received))

    #json_recv()
```
